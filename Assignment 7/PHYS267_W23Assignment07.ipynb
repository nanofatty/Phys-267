You perform three trials of an experiment, where you take a measurement of absorbance every minute from 1 minute to 80 minutes. Use the data set A7_Q4Data.txt and answer the following questions.

#### Solution

**Part 1 --------------------------------------------**

````python
# Loading Data
data = np.loadtxt("A7_Q4Data.txt", skiprows=1)

# Trials extraction
time = data[:,0]
trial1 = data[:, 1]
trial2 = data[:, 2]
trial3 = data[:, 3]

# Plots, labels and legend
fig4, ax4 = plt.subplots()
ax4.plot(trial1, label="Trial 1")
ax4.plot(trial2, label="Trial 2")
ax4.plot(trial3, label="Trial 3")
ax4.set_xlabel("Time (min)")
ax4.set_ylabel("Absorbance")
ax4.legend()
plt.show()
````

**Part 2 --------------------------------------------**

````python
from scipy.stats import skew, kurtosis
# Averaging abs values across trials
abs_avg = np.mean(data[:,1:], axis=1)
abs_std = np.std(data[:,1:], axis=1, ddof=2)
# Moments of averaged values
mean = np.mean(abs_avg)
variance = np.var(abs_avg, ddof=2)
skewness = skew(abs_avg)
kurt = kurtosis(abs_avg)

print("Mean:", mean)
print("Variance:", variance)
print("Skewness:", skewness)
print("Kurtosis:", kurt)
````

**Part 3 --------------------------------------------**

````python
fig5, ax5 = plt.subplots(figsize=(45,15))

ax5.errorbar(time, abs_avg, yerr=abs_std, capsize=10, label='Error', fmt='o', markersize=1)
ax5.plot(time, abs_avg, '-', linewidth=1, label='Average', color='red')
ax5.set_xlabel('Time (min)')
ax5.set_ylabel('Absorbance')
ax5.set_title('Average Absorbance Measurements with Error Bars')
ax5.legend()
plt.show()
````
