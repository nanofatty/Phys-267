### <font color="blue">2.2 Analyzing Qualitative Data</font>

When we have qualitative data, we already know what values of the variable have been measured. Qualitative data is synonymous to **categorical** data, so we are most likely going to present this data in the form of categories and how frequent each category is (or how often that category occurs). 

#### 2.2.1 Types of Categorical Data

Of the categorical datatypes, we can further split this into three types of values:
1. **Boolean** data which only have two possible values: true/false, weekdays/weekends, etc.
2. **Nominal** data that requires more than two categories: married/single/divorced, undergraduate/graduate/postgraduate studies, etc.
3. **Ordinal** data which are ordered and have logical sequence: strongly disagree, disagree, neutral, agree, strongly agree for example

#### 2.2.2 Frequency, Relative Frequency and Percentage

When we have qualitative variables, it is recommended that we create a **statistical table** that lists out all of the categories along with a count (measure) of how often each value occurred. But how do we quantify "how often"?
- We can use **frequency** which represents the number of measurements in each category (ie. 45 events happened in this category)
- We can use **relative frequency** which gives us the proportion of measurements in each category (ie. 45/200 events happened in this category = 0.225)
- We can use **percentage** which gives us the % of measurements in each category (ie. 22.5% of events were in the category)

Now, how are these terms related to each other? Assuming $n$ is the total number of measurements in your set:

$$\text{relative frequency} = \frac{\text{frequency}}{n} \qquad \text{percentage} = 100 \times \text{relative frequency}$$

Note the following:
- The sum of all frequencies will be $n$
- The sum of all relative frequencies will be 1
- The sum of all percentages will be 100 (or 100%)

Prior to even creating a visualization of your categorical data, we must first be able to determine the best categories to filter our data into. When creating categories, keep in mind these two key characteristics:
- A measurement must fall into one of your chosen categories (always include an "other" for all outliers)
- A measurement must fall into only 1 category

### <font color="blue">2.3 Plots for Categorical Data</font>

Visualization is critical in data analysis, especially for numerical data. When it comes to categorical data, the most important part of visualization is to see the frequency of each category. Sometimes, this may just come in the form of a statistical table that presents the categories alongside some statistical metric (frequency, relative frequency, percentage). Other times, we want to present this in a graph such as a pie chart or bar chart. 

#### 2.3.1 Pie Chart

Let's walk through an example together of how we can use pie charts to represent categorical data. In astronomy, we can classify stars based on their spectral characteristics. We will use the following data from the *Harvard System of Spectral Classification* obtained from Glenn Ledrew (2001, <a href="https://articles.adsabs.harvard.edu/pdf/2001JRASC..95...32L">"The Real Starry Sky"</a>) and compiled in Wikipedia:

```python
import pandas as pd   # Don't worry about pandas yet; we will have a more detailed overview in later chapters
stellarclassification = [["O", 0.00003], ["B", 0.13], ["A", 0.6], \
                         ["F", 3.0], ["G", 7.6], ["K", 12.1], ["M", 76.45], \
                         ["Other", 0.11997]]
pd.DataFrame(stellarclassification, columns=["Class", "Percentage"])
```

A **pie chart** is very useful when it comes to seeing the frequency distribution of categories. Our goal is to take the stellar classification data above and present it as a pie chart. The Python packages we need will be:
- `matplotlib.pyplot` imported as `plt`for short
- `plt.pie` to plot a pie chart

Apart from that, we will try the following:
- label the slices using `labels`
- label the percentages using `autopct`
- offset any chosen slice using `explode`
- start the pie at any angle using `startangle`
- creat a legend using `legend()`
- create visual effects (shadows and 3D effect)

```python
import matplotlib.pyplot as plt
import numpy as np

# Create pie chart in counter-clockwise fashion
stellarclasses = 'O', 'B', 'A', 'F', 'G', 'K', 'M', 'Other'
stellarslices = [0.00003, 0.13, 0.6, 3, 7.6, 12.1, 76.45, 0.11997]
stellaroffset = (0, 0, 0, 0, 0, 0, 0.7, 0)
stellarcolours = ['darkblue', 'blue', 'cornflowerblue', 'ghostwhite', 'lightyellow', 'navajowhite', 'darkorange', 'black']
plt.pie(stellarslices, explode=stellaroffset, labels=stellarclasses, \
        colors=stellarcolours, autopct='%1.1f%%', shadow=True, startangle=100)
plt.axis('equal')
plt.title("Stellar Classification of Stars", pad=20, fontsize=18, color='darkblue')
plt.show()
```

As you may notice with the example above, there is overlap in the labels, and you can barely see some of the wedges. Based on this, there are some options:
- Remove the labels and add in a legend to clean up the pie chart
- Use a different type of graph to represent this information

```python
# We don't have to import again if we run the code cell from above with the imported packages
plt.pie(stellarslices, explode=stellaroffset, labels=None, \
        colors=stellarcolours, autopct='%1.0f%%', shadow=True, startangle=100)
plt.legend(labels=stellarclasses, title="Classes", loc='center left', bbox_to_anchor=(1, 0, 0.2, 1))
plt.axis('equal')
plt.title("Stellar Classification of Stars", pad=20, fontsize=18, color='darkblue')
plt.show()
```

This doesn't really solve the how crammed the labels are in the pie chart, does it? Let's see if there's another way to do it by creating labels with arrows pointing to each section. This means we need to place these labels properly around the pie chart. Since it is a circle, we can use angles to define these positions.

```python
# We need to define the wedges and the text involved
# To do this more easily, we will use object-oriented programming 
# which will be covered later on, where you define a figure and axis first
stellarfig, ax = plt.subplots(figsize=(6, 5), subplot_kw=dict(aspect="equal"))
stellarlabels = ["0.00003%", "0.13%", "0.6%", "3%", "7.6%", "12.1%", "76.45%", "0.11997%"]
stellarslices = [0.00003, 0.13, 0.6, 3, 7.6, 12.1, 76.45, 0.11997]
# Try changing the width in "wedgeprops" to see what happens - note that you can only go from 0-1
stellarwedges, texts = ax.pie(stellarslices, wedgeprops=dict(width=1), startangle=-125)

# Define label's border box
bbox_props = dict(boxstyle="square,pad=0.5", fc="w", ec="k", lw=0.3)
kw = dict(arrowprops=dict(arrowstyle="-"), bbox=bbox_props, zorder=0, va="center")

# Create labels
for i, p in enumerate(stellarwedges):
    angle = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(angle))
    x = np.cos(np.deg2rad(angle))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(angle)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    ax.annotate(stellarlabels[i], xy=(x, y), xytext=(x*i/1.15, 1.7*y),
                horizontalalignment=horizontalalignment, **kw)

ax.legend(labels=stellarclasses, title="Classes", loc='center left', bbox_to_anchor=(1.5, 0, 20, 1))
plt.show()
```

Is this better? We notice there are too many lines now that overlap. We could probably play around with the code to adjust the lengths of the lines, but at some point, the wedges are so tiny, you can't even tell them apart. In this case, it may be good to omit this data or add it to the "Other" category so you can get a better graph. Try it out yoursefl!

Are there other ways in Python to present pie charts? Yes! There is never just "one solution", but many. We started off using `pandas` to display a table. Logically, the `pandas` package should allow us to present a pie chart as well. To do this, we would simply have our data presented as follows:

```python
import pandas as pd
stellarclassification = pd.DataFrame({'Frequency': [0.00003, 0.13, 0.6, 3.0, 7.6, 12.1, 76.45]}, \
                                     index=["O", "B", "A", "F", "G", "K", "M"])
plot = stellarclassification.plot.pie(y='Frequency', figsize=(5,5))
```

#### 2.3.2 Bar Chart

A **bar chart** is another common graph used to display categorical data. We pick one axis to be our categories and the other to represent frequency, relative frequency or percentage. In most cases, it would be *frequency*.

Our goal is to take the stellar classification of stars from the previous example and represent that as a bar chart. This time, however, let's use the frequency data from its source instead of percentage.

```python
import pandas as pd
stellarclassification = [["O", 0.00025], ["B", 1], ["A", 5], \
                         ["F", 25], ["G", 63], ["K", 100], ["M", 630]]
pd.DataFrame(stellarclassification, columns=["Class", "Frequency"])
```

The Python code we will need to create a bar chart will be:
- `matplotlib.pyplot` imported as `plt`
- `numpy` as `np` to store values
- `plt.bar` to display our x and y axes

You should be able to do the following:
- Present the bar chart vertically or horizontally
- Label the bar chart axes
- Change the width and colour of the bar

```python
import matplotlib.pyplot as plt
import numpy as np

barclass = np.array(["O", "B", "A", "F", "G", "K", "M"])
barfrequency = np.array([0.00025, 1, 5, 25, 63, 100, 630])

plt.bar(barclass, barfrequency, color='green', width=0.3)
plt.title("Vertical Bar Chart")
plt.xlabel("Stellar Classes of Stars")
plt.ylabel("Frequency")
plt.show()
```

To create a horizontal bar chart, use `plt.barh` instead of `plt.bar`:

```python
plt.barh(barclass, barfrequency, color='darkblue', height=0.8)
plt.title("Horizontal Bar Chart")
plt.xlabel("Frequency")
plt.ylabel("Stellar Classes of Stars")
plt.show()
```

Can we convert either of these into a 3D plot? Let's try using another feature of `plt` called `projection='3d'`!

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(10, figsize=(12,8))    # creates a new figure and number (recommended)
ax = fig.add_subplot(projection='3d')   # creates a 3d coordinate system

color = 'purple'
barclass = np.array(["O", "B", "A", "F", "G", "K", "M"])
barfrequency = np.array([0.00025, 1, 5, 25, 63, 100, 630])

plt.bar(barclass, barfrequency, zdir='y', color=color, alpha=0.8, width=0.7)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

### 2.4 Proper Creation of Figures

You may have noticed that we did something a bit different when we created the 3D plot. We used **`fig = plt.figure()`**. Why is that? And why is this recommended? As physicits, we want to be able to customize our graphs and this customization requires us to think in an *object-oriented* fashion. This means that when we think about plotting a graph, we are doing the following instead of just sticking with the 'default' plots created by `matplotlib`:
1. Create a figure object (like a canvas) of a certain size using `fig = plt.figure()`
    - Put `figsize=(2,6)` inside the brackets to create a figure with that particular width and height respectively
    - This is quite important for publications as you want the image to be static instead of having it be automatically adjusted through default settings
        
    
2. Create an axis object within the figure by using `ax = fig.add_subplot(1,1,1)`
    - This will create a 1x1 grid with the 3rd value being the ID of the subplot
    
3. Optional: Create additional axes within the figure by using `ax2 = fig.add_axes()`
    - This will create a figure within a figure in case you want to compare or zoom in to certain parts of your data
    
    
4. Plot your points using `ax.plot()` making sure you use `ax1` or `ax2` to plot the corresponding axes


5. Use `ax.set_ylabel()` to label your y-axis and `ax.set_xlabel()` to label your x-axis

**You can combine Steps 1 and 2 using `figure_name, axis_name = plt.subplots()` which creates a figure that contains a single axis; then use `axis_name.plot()` to plot data on that axis_name**

```python
fig = plt.figure(figsize=(10,6))
ax1 = fig.add_subplot(1,1,1)
ax2 = fig.add_axes([0.58, 0.55, 0.3, 0.3])

# Plot on Axis 1 (larger one) with random data
t = np.linspace(0, 10, 100)
ax1.plot(t, 1/(t**2+1), 'b-')
ax1.set_ylabel("Blue Curve Y-Values")
ax1.set_xlabel("Blue Curve X-Values")
ax2.plot(t, t, 'g-')
ax2.set_ylabel("Green Curve Y-Values")
ax2.set_xlabel("Green Curve X-Values")
```
