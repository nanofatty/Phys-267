#### 3.2.3 Line Charts

In Physics, we regularly need to collect data over equally spaced intervals to create a **time series**, a data set that is  based on time. These are most effectively presented as **line charts**. From these line charts, our goal is to determine a pattern or **trend** that will likely continue into the future, and then use that pattern to make a prediction. We will not cover this yet, but rather once we reach the Data Analysis portion of this course when we need to model a set of data.

Once again, we will be using `matplotlib.pyplot` to create our line charts, specifically the function `plot(x,y)`

```python
import matplotlib.pyplot as plt
import numpy as np

# Create data
time = np.linspace(0, 10, 50)  # Create 50 data points from 0 to 10
y = np.sin(time)               # Create 50 y-points based on x-points

fig, ax = plt.subplots()    # Create figure and axis
ax.plot(time, y, 'b.')         # Plot points using blue dots
ax.plot(time, y, 'r-')         # Plot red line through the points
plt.xlabel("Time")
plt.ylabel("Y-Values")
plt.show()
```

When working with line charts, there are additional parameters we can customize for the `plot` function:
- **Marker**: this is the "dot" that represents your data point, set by `marker=`
    - `.` for a point
    - `,` for a pixel
    - `o` for a circle
    - `v`, `^`, `<`, `>`, for a downwards / upwards / left / right triangle
    - `s` for a square
    - `*` for a star
    - `x` for an x, or `X` for a large, filled X
    - `d` for a thin diamond and `D` for a thick diamond
    - `+` for a thin plus sign and `P` for a thick one
    

- **Marker Size**: we can select the `markersize=` in points

    
- **Line Style**: this is how the line looks if you choose a line based on `linestyle=`
    - `-` for solid line
    - `--` for dashed line
    - `-.` for a dashed/dotted line
    - `:` for a dotted line
    

- **Line Width**: determines how thick your line is based on `linewidth=`

Let's try to play around with these settings through some examples below.

```python
fig, ax = plt.subplots(figsize=(10,6))
ax.plot(time, y, marker='o', markersize=8, markerfacecolor='red', \
        markeredgecolor='black', linestyle=':', linewidth=3, \
        color='royalblue')
plt.xlabel("Time")
plt.ylabel("Y-Values")
plt.show()
```

#### 3.2.4 Histograms

Similar to bar charts, **histograms** "bin" data into numerical categories. But, when would we use this? We could regularly use this for grouping together grades (ie. GPAs, test scores, etc.) so that we can see what portion of the students fall into which grade-ranges. 

Specifically, we want to take a look at **relative frequency histograms**, which allow us to determine the distribution of a set of data in terms of its *location* and *shape*. If you have a large set of data, you can follow these steps to create a relative frequency histogram:
- Choose the number of classes your data falls within (usually between 5-15)
- Calculate the width of each class by dividing the difference between the largest and smallest values by the number of classes you've chosen
- Round your class width to a convenient value (ie. ages 1-4, 5-9, etc.)
- Create a statistical table with the classes, their frequencies, and their relative frequencies
- Create the histogram

To test this out, we can generate some random, dummy data and use `plt.hist()` or `ax.hist()`.

```python
import numpy as np
import matplotlib.pyplot as plt
import random

# Make dummy data based on normal distribution (will cover in later units)
x = random.sample(range(0, 100), 50)

# Generate plot
fig, ax = plt.subplots()
ax.hist (x, bins=12, edgecolor='black')
ax.set_title('Simple Histogram')
plt.show()
```

```python
# What about a relative frequency histogram? We can use matplotlib or scipy.
# Return evenly-spaced values within a given interval
bins = np.arange(0, 101, 10)    
# Define histogram values and where the edges are (bins)
hist, edges = np.histogram(x, bins)
# Define rel. freq. as x / total values (hist.sum())
relfreq = hist/float(hist.sum())
# Plot a bar graph with bin values (x-axis), relfreq (y-axis)
plt.bar(bins[:-1], relfreq, width=8, align='center', color='green')
# Show the plot
plt.title("Relative Frequency Histogram")
plt.show()
```

```python
# Make dummy data for three sets of data
np.random.seed(1)
x = np.random.randn(1000, 3)
colors = ['blue', 'orange', 'green']

# Generate plot with multiple sets of data
fig, ax = plt.subplots()
ax.hist (x, bins=12, edgecolor='black', label=colors)
ax.set_title('Multiple Sets of Data')
ax.legend()
plt.show()
```

#### 3.2.5 Interpreting Graphs

After looking at creating graphs, we need to be aware of how to read them. Here are some key terms and definitions you should be aware of when discussing the characteristics of graphs:

- **Scales** define the values of what is being measured
- **Location** refers to the centre of a distribution
- **Shape** refers to its shape and whether there are peaks involved
- **Outlier** is a measurement that is much bigger/smaller than all others (see below for discussion on standard deviations and what can be considered a likely outlier)

Meanwhile, a **distribution** is commonly defined by its shape:
- **Symmetric Distributions**: left and right sides are mirror images when split down the middle
- **Skewed to the Right**: greater portion of measurements are on the right side of the peak value
- **Skewed to the Left**: greater portion of measurements lie to the left of the peak value
- **Unimodal Distribution**: only has one peak
- **Bimodal Distribution**: has two peaks (represents two different populations in the data set)

### <font color="blue">3.3 Numerical Measures of Quantitative Data</font>

Although graphs are useful for visualizing a data set, they are not always the greatest tool when you want to make inferences about a population from your sample. As a result, we must be able to use **numerical measures** to analyze our data.

Using our data set, we will be calculing a set of *numbers* that help convey a good mental picture of the frequency distribution. These measures are called **parameters** when we associate them with the population, and they are called **statistics** when we calculate them from the sample measurements.

Throughout the next few sections, we will be covering measures of **centre**, **spread** and **relative standing**. Additional graphs will be presented if relevant to the section.

#### 3.3.1 Measures of Centre

If we have a set of data that is spread across one axis (like the x-axis), we can determine the centre of this distribution through various methods: mean, median and mode. These are terms that we should all be quite familiar with already!

<div class="alert alert-block alert-success">
    <p><b>Mean (Average)</b>: From a set of $n$ measurements, this is equal to the sum of all measurements divided by $n$</p>
    <p>Assuming $x_i$ is the $i$th measurement of variable $X$, we can define the mean as follows:</p>
    <p>$$\bar{x} = \frac{\sum{x_i}}{n}$$</p>
</div>

To determine the mean of a set of data, we can use `np.mean()`

```python
import numpy as np
velocity = [10, 20, 25, 78, 43, 190, 243, \
            11, 5, 5, 0, 100, 72, 5, 47]
velocity_avg = np.mean(velocity)
print("The average velocity is:", velocity_avg)
```

<div class="alert alert-block alert-success">
    <p><b>Median</b>: From a set of $n$ measurements, the median $m$ is the value of $x$ that falls in the middle position when all measurements are ordered from smallest to largest.</p>
</div>

To determine the median of a set of data, let's try `np.median()`

```python
velocity_mdn = np.median(velocity)
print("The median velocity is:", velocity_mdn)
```

<div class="alert alert-block alert-success">
    <p><b>Mode</b>: From a set of $n$ measurements, the mode is the most frequently occurring value of $x$ or the most frequent category that occurs. If the mode is within a class of measurements that have been binned together (called a <i>modal class</i>), the midpoint of that class is taken to be the mode.</p>
</div>

If we want to determine the mode of a set of data, we need to use `stats.mode()` from `scipy`, and get the value of the most common value (mode=array) as well as its frequency (count=array):

```python
from scipy import stats
velocity_mode = stats.mode(velocity)
print("The mode of velocity is:", velocity_mode)
```

#### 3.3.2 Measures of Spread

Another important aspect of your data set would be its spread, how far that data spreads out from the centre of the distribution. In statistical terms, we call this **variability**, **dispersion**, or **spread** a very important characteritsic of data that can be measured as a range, variance, and standard deviation.

<div class="alert alert-block alert-success">
    <p><b>Range</b>: The range, or $R$, of a set of $n$ measurements is defined as the difference between the largest and smallest measurements. This gives us a good idea of the spread of our data.</p>
</div>

The range is something we can find easily be determining the minimum `min()` and maximum `max()` of a list of values.

```python
velocity_max = max(velocity)
velocity_min = min(velocity)
print("The range of velocity is from", velocity_min, "to", velocity_max)
```

When we look at our measurements, each measurement is a certain distance away from the mean. We call that distance the *deviation*. Measurements to the right of the mean produce positive deviations, while those to the left create negative deviations. However, these deviations, if summed directly, will always add up to zero (because you always have an equal amount of deviations to the left/right of your average). As a result, we need to find another way to represent this deviation. 

We now introduce the **sum of squares**. We square the deviations and sum them up to calculate what we call a **variance**. Note that the variance is different for a population than for a sample.

<div class="alert alert-block alert-success">
    <p><b>Variance for Population</b>: For a population of $N$ measurements, the population variance is represented by $\sigma^2$ and is the average of the squares of the deviations of the measurements around their mean $\mu$:</p>
    <p>$$\sigma^2=\frac{\sum{(x_i - \mu)^2}}{N}$$</p>
</div>

Usually, we have to calculate the variance of a sample of $n$ measurements...

<div class="alert alert-block alert-success">
    <p><b>Variance for Sample</b>: For a sample of $n$ measurements, the sample variance is represented by $s^2$ and is the sum of the squares of the deviations of the measurements around their mean $\bar{x}$, divided by $n$-1:</p>
    <p>$$s^2=\frac{\sum{(x_i - \bar{x})^2}}{n-1}$$</p>
</div>

If we take the square root of this variance, we obtain a **standard deviation**.

<div class="alert alert-block alert-success">
    <p><b>Standard Deviation</b>: The positive square root of the variance, denoted as $\sigma$ for population or $s$ for sample.</p>
</div>

Now, why do we have to divide by $n-1$ for the sample variance? It is for the simple reason that the sample variance $s^2$ gives better estimates of $\sigma^2$ with $n-1$ than with just $n$; so we always divide by $n-1$ when computing the sample variance and sample standard deviation. 

Now that we have gone through some numerical measures of spread, keep in mind the following:
- The value of $s$ is always greater than or equal to zero
- The larger the value of $s^2$ or $s$, the greater the variability of the data set
- If $s^2$ or $s$ is equal to zero, all the measurements must have the same value
- If we want to measure the variability of our sample in the same units as our original observations, we simply calculate the standard deviation $s = \sqrt{s^2}$

#### 3.3.3 Measures of Relative Standing

We have now covered ways of analyzing the centres and spread of a distribution, but what about comparisons within a distribution? Sometimes, we may want to know how one observation is relative to another within our dataset. We can use one of several methods to analyze this: z-scores, percentiles, and quartiles.

Let's start with the **z-score** for a sample. 

<div class="alert alert-block alert-success">
    <p><b>Z-Score</b>: Measures the distance between an observation and the mean, in units of standard deviation</p>
    <p>$$\text{z-score}=\frac{x-\bar{x}}{s}$$</p>
</div>

Let's say you get a z-score of 1.3. This means that your observation lies 1.3 standard deviations above the mean.

This z-score is a very reliable tool when it comes to determining whether your observation is an **outlier** or not. Based on the Empiral Rule and Tchebysheff's Theorem:
- At least 75% and more likely 95% of observations will lie within 2 standard deviations of their mean (-2 to 2) and thus anything that exceeds this is somewhat unlikely to take place (~5%)
- At least 89% and more likely 99.7% of observations lie within 3 standard deviations of their mean (-3 to 3) and thus anything that exceeds this is very unlikely to take place (<1%)

The **percentile** is another meausre of relative standing that is often used for large data sets (like performance amongst a school cohort). 

<div class="alert alert-block alert-success">
    <p><b>Percentile</b>: After arranging a set of $n$ measurements for a variable $x$ based on order of magnitude, the <b>$p$th percentile</b> is the value of $x$ that is greater than $p$% of the measurements and is less than the remaining $100-p$%)</p>
</div>

What does this mean exactly? It means that if you score at the 70th percentile, 70% of all other values are lower than your score, and 30% were higher. 

Within this, we can also define the following:
- **lower quartile** or first quartile, is the value of $x$ that is greater than 25% of the measurements and less than the remaining 75% of measurements
- **upper quartile** or third quartile is the value of $x$ that is greater than 75% of the measurements and less than the remaining 25% of measurements
- this leaves the **second quartile**, called the median because it is the value of $x$ that is greater than 50% of measurements and less than 50% of measurements.
- the **interquartile range (IQR)** for a set of measurements is the difference between the upper and lower quartiles

For now, knowing the definitions is key. We will be going through programming syntax for various statistical measures once we start working with datasets in the data analysis chapters later this term.

```python
mean_RanNums2 = np.mean(randNums2)

print("Mean should be the same as the last one", mean_RanNums2)

#Sum of Squares
sum_squares= 0
for i in range(len(randNums2)):
    dev = i  - mean_RanNums2
    square = dev**2
    sum_squares += square

print("Sum of squares", sum_squares) 

print("Variance", sum_squares/(len(randNums2)-1))

print("std:", np.sqrt(sum_squares/(len(randNums2)-1)))

```
